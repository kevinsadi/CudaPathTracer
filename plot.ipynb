```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# parallel runtime with spp
spp = np.array([128, 256, 512, 1024, 2048, 4096])
cpu_time = np.array([14.741, 19.598, 31.638, 53.186,95.684, 173.334])
gpu_time = np.array([1.387, 2.76,5.522,11.008,21.998,43.968])

font = 20

plt.figure(figsize=(20, 10))
plt.subplot(1, 1, 1)
plt.plot(spp, cpu_time, label="CPU")
plt.scatter(spp, cpu_time)
plt.plot(spp, gpu_time, label="GPU")
plt.scatter(spp, gpu_time)
plt.xticks(spp)
plt.yticks(np.concatenate((cpu_time, gpu_time)))
plt.xlabel("SPP", fontsize=font)
plt.ylabel("Time(second)", fontsize=font)
plt.legend()
plt.show()
```

```python
# strong scaling with num of threads
num_threads = np.array([2, 4, 8, 16, 32, 64])
cpu_time = np.array([303.107, 156.765, 82.981, 45.554, 27.656, 24.619])
gpu_time = np.array([19.221, 13.088, 8.511, 5.363, 3.241, 3.249])

font = 20

plt.figure(figsize=(20, 10))
plt.subplot(1, 1, 1)
plt.plot(num_threads, cpu_time, label="CPU")
plt.scatter(num_threads, cpu_time)
plt.plot(num_threads, gpu_time, label="GPU")
plt.scatter(num_threads, gpu_time)
plt.xticks(num_threads)
plt.yticks(np.concatenate((cpu_time, gpu_time)))
plt.xlabel("Num Threads", fontsize=font)
plt.ylabel("Time(second)", fontsize=font)
plt.legend()
plt.show()
```

```python
# strong scaling speedup
num_threads = np.array([2, 4, 8, 16, 32, 64])
cpu_speedup = cpu_time[0] / cpu_time
gpu_speedup = gpu_time[0] / gpu_time

font = 20

plt.figure(figsize=(20, 10))
plt.subplot(1, 1, 1)
plt.plot(num_threads, cpu_speedup, label="CPU")
plt.scatter(num_threads, cpu_speedup)
plt.plot(num_threads, gpu_speedup, label="GPU")
plt.scatter(num_threads, gpu_speedup)
plt.xticks(num_threads)
plt.yticks(np.concatenate((cpu_speedup, gpu_speedup)))
plt.xlabel("Num Threads", fontsize=font)
plt.ylabel("Speedup", fontsize=font)
plt.legend()
plt.show()
```

```python
# weak scaling
num_threads = np.array([2, 4, 8, 16, 32, 64])
cpu_time = np.array([132.776,134.969,138.557,144.012,155.104,173.751])
gpu_time = np.array([8.218, 11.195,14.539,18.235,21.976,43.987])

font = 20

plt.figure(figsize=(20, 10))
plt.subplot(1, 1, 1)
plt.plot(num_threads, cpu_time, label="CPU")
plt.scatter(num_threads, cpu_time)
plt.plot(num_threads, gpu_time, label="GPU")
plt.scatter(num_threads, gpu_time)
plt.xticks(num_threads)
plt.yticks(np.concatenate((cpu_time, gpu_time)))
plt.xlabel("Num Threads", fontsize=font)
plt.ylabel("Time(second)", fontsize=font)
plt.legend()
plt.show()
```

```python
# weak scaling efficiency
cpu_efficiency = cpu_time[0] / (cpu_time * num_threads)
cpu_efficiency = cpu_efficiency[0] / cpu_efficiency[:]
cpu_scaled_speedup = cpu_efficiency * num_threads

gpu_efficiency = gpu_time[0] / (gpu_time * num_threads)
gpu_efficiency = gpu_efficiency[0] / gpu_efficiency[:]
gpu_scaled_speedup = gpu_efficiency * num_threads

plt.figure(figsize=(20, 10))
plt.subplot(1, 1, 1)
plt.plot(num_threads, cpu_efficiency, label="CPU")
plt.scatter(num_threads, cpu_efficiency)
plt.plot(num_threads, gpu_efficiency, label="GPU")
plt.scatter(num_threads, gpu_efficiency)
plt.xticks(num_threads)
plt.yticks(np.concatenate((cpu_efficiency, gpu_efficiency)))
plt.xlabel("Number of Threads", fontsize=font)
plt.ylabel("Efficiency", fontsize=font)
plt.legend()
plt.show()
```

```python
# weak scaling scaled speedup
cpu_efficiency = cpu_time[0] / (cpu_time * num_threads)
cpu_efficiency = cpu_efficiency[0] / cpu_efficiency[:]
cpu_scaled_speedup = cpu_efficiency * num_threads

gpu_efficiency = gpu_time[0] / (gpu_time * num_threads)
gpu_efficiency = gpu_efficiency[0] / gpu_efficiency[:]
gpu_scaled_speedup = gpu_efficiency * num_threads

plt.figure(figsize=(20, 10))
plt.subplot(1, 1, 1)
plt.plot(num_threads, cpu_scaled_speedup, label="CPU")
plt.scatter(num_threads, cpu_scaled_speedup)
plt.plot(num_threads, gpu_scaled_speedup, label="GPU")
plt.scatter(num_threads, gpu_scaled_speedup)
plt.xticks(num_threads)
plt.yticks(np.concatenate((cpu_scaled_speedup, gpu_scaled_speedup)))
plt.xlabel("Number of Threads", fontsize=font)
plt.ylabel("Scaled Speedup", fontsize=font)
plt.legend()
plt.show()
```
